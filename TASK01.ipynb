```python
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_csv('/content/population.csv', skiprows=4)
year = '2020'
year_data = data[year].dropna()
print("ðŸ“Œ Loaded data successfully")

print("ðŸ“Œ Showing histogram...")
plt.figure(figsize=(10,6))
plt.hist(year_data, bins=30, color='teal', edgecolor='black')
plt.title(f'Histogram: Global Population Distribition({year})')
plt.xlabel('population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.savefig('population_histogram.png')
plt.show()


top_10 = data[['Country Name', year]].dropna().sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_10['Country Name'], top_10[year], color='purple')
plt.title(f'Top 10 most populated countries in {year}')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top10_bar_chart.png')
print("ðŸ“Œ Showing bar chart...")
plt.show()

print("ðŸ“Œ Showing statistics...")
print(f"\n Population Statistics for {year}")
print(f"Minimum: {year_data.min():,.0f}")
print(f"Maximum: {year_data.max():,.0f}")
print(f"Average: {year_data.mean():,.0f}")
```
